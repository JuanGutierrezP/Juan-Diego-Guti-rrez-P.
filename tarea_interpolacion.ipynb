```python
#Punto 3 interpolación

import numpy as np
import matplotlib.pyplot as plt
x0=100
x1=121
x2=144
def f_inter(x):
   return np.sqrt(x)
y0 = f_inter(x0)
y1 = f_inter(x1)
y2 = f_inter(x2)
# x a evaluar
x = 114
#coef de lagrange sin función
L0 = (x- x1)*(x-x2) / ((x0- x1)*(x0-x2))
L1 = (x- x0)*(x-x2) / ((x1- x0)*(x1-x2))
L2 = (x- x0)*(x-x1) / ((x2- x0)*(x2-x1))

y_casi = L0 * y0 + L1 * y1 + L2 * y2

error_abs = np.abs(f_inter(x) - y_casi)



print(error_abs, f_inter(114), y_casi)

```

```python
# Punto 4

import numpy as np
import pandas as pd
from scipy.interpolate import lagrange
import matplotlib.pyplot as plt

# Paso 1: Cargar los datos desde el archivo CSV
url = 'https://raw.githubusercontent.com/asegura4488/Database/main/MetodosComputacionalesReforma/Parabolico.csv'
data = pd.read_csv(url)

# Mostrar los primeros datos para asegurarnos de que se ha cargado correctamente
print(data.head())

# Paso 2: Realizar la interpolación de Lagrange
# Extraer las columnas de posición X e Y
x = data['X'].values
y = data['Y'].values

# Usar la interpolación de Lagrange
polynomial = lagrange(x, y)

# Mostrar el polinomio obtenido
print(f'Polinomio interpolado: {polynomial}')

# Paso 3: Graficar la trayectoria interpolada
x_vals = np.linspace(min(x), max(x), 100)
y_vals = polynomial(x_vals)

plt.plot(x, y, 'ro', label='Datos originales')
plt.plot(x_vals, y_vals, label='Interpolación de Lagrange')
plt.title('Interpolación de la trayectoria de la bala')
plt.xlabel('Posición X (m)')
plt.ylabel('Posición Y (m)')
plt.legend()
plt.grid(True)
plt.show()

# Paso 4: Extraer los coeficientes cuadráticos y lineales del polinomio
coef_quadratic = polynomial.coef[-3] if len(polynomial.coef) >= 3 else 0
coef_linear = polynomial.coef[-2] if len(polynomial.coef) >= 2 else 0

print(f'Coeficiente cuadrático (t^2): {coef_quadratic}')
print(f'Coeficiente lineal (t): {coef_linear}')
```
