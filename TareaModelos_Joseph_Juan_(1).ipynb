```python
#Problema 1

# Datos proporcionados
total_participantes = 1000

# Número de hombres y mujeres
hombres = 600
mujeres = 400

gafas_hombres = 185
gafas_mujeres = 115
gafas_total = gafas_hombres + gafas_mujeres  # 300

# a) Probabilidad de ser hombre P(H)
P_H = hombres / total_participantes
print(f"a) Probabilidad de ser hombre P(H) = {hombres}/{total_participantes} = {P_H:.4f}")

# b) Probabilidad de ser mujer P(M)
P_M = mujeres / total_participantes
print(f"b) Probabilidad de ser mujer P(M) = {mujeres}/{total_participantes} = {P_M:.4f}")

# c) Probabilidad de usar gafas P(G)
P_G = gafas_total / total_participantes
print(f"c) Probabilidad de usar gafas P(G) = {gafas_total}/{total_participantes} = {P_G:.4f}")

# d) Probabilidad de que una mujer use gafas P(G|M)
# P(G ∩ M) = mujeres que usan gafas / total participantes
P_G_and_M = gafas_mujeres / total_participantes

# P(M) ya calculado
# P(G|M) = P(G ∩ M) / P(M)
P_G_given_M = P_G_and_M / P_M
print(f"d) Probabilidad de que una mujer use gafas P(G|M) = {P_G_and_M:.4f}/{P_M:.4f} = {P_G_given_M:.4f}")
```

```python
#Problema 3

import numpy as np

# Número de simulaciones
N = 1000000

# Contador de casos favorables
count_FF = 0

for _ in range(N):
    # Inicializamos la bolsa
    bag = ['L'] * 2 + ['F'] * 3
    # Primer dulce
    first_candy = np.random.choice(bag)
    bag.remove(first_candy)
    # Segundo dulce
    second_candy = np.random.choice(bag)
    # Verificamos si ambos son de fresa
    if first_candy == 'F' and second_candy == 'F':
        count_FF += 1

# Calculamos la probabilidad
P_FF_sim = count_FF / N

print(f"P(F ∩ F) simulada = {P_FF_sim:.4f} (Teórica: 0.3000)")
```

# Teorema de Bayes

```python
# Problema 1

# Probabilidad de que un paciente fume
P_F = 0.4

# Probabilidad de que un paciente sea hombre dado que fuma
P_H_given_F = 0.75

# Probabilidad de que un paciente sea mujer dado que fuma
P_M_given_F = 1 - P_H_given_F  # 0.25

# Probabilidad de que un paciente no fume
P_not_F = 1 - P_F  # 0.6

# Probabilidad de que un paciente sea mujer dado que no fuma
P_M_given_not_F = 0.6

# Probabilidad de que un paciente sea hombre dado que no fuma
P_H_given_not_F = 1 - P_M_given_not_F  # 0.4

# a) Calcular P(M) (probabilidad de que el paciente sea mujer)
# P(M) = P(M ∩ F) + P(M ∩ F')

# P(M ∩ F) = P(F) * P(M | F)
P_M_and_F = P_F * P_M_given_F  # 0.4 * 0.25 = 0.1

# P(M ∩ F') = P(F') * P(M | F')
P_M_and_not_F = P_not_F * P_M_given_not_F  # 0.6 * 0.6 = 0.36

# P(M) = P(M ∩ F) + P(M ∩ F')
P_M = P_M_and_F + P_M_and_not_F  # 0.1 + 0.36 = 0.46

print("a) La probabilidad de que el paciente sea mujer es:", P_M)

# b) Calcular P(H ∩ F) (probabilidad de que el paciente sea hombre y fume)
# P(H ∩ F) = P(F) * P(H | F)
P_H_and_F = P_F * P_H_given_F  # 0.4 * 0.75 = 0.3

print("b) La probabilidad de que el paciente sea hombre y fume es:", P_H_and_F)

# c) Calcular P(F | M) (probabilidad de que el paciente fume dado que es mujer)
# P(F | M) = P(M ∩ F) / P(M)
P_F_given_M = P_M_and_F / P_M  # 0.1 / 0.46

print("c) La probabilidad de que el paciente fume dado que es mujer es:", P_F_given_M)

# Expresar los resultados como fracciones
from fractions import Fraction

P_M_fraction = Fraction(P_M).limit_denominator()
P_H_and_F_fraction = Fraction(P_H_and_F).limit_denominator()
P_F_given_M_fraction = Fraction(P_F_given_M).limit_denominator()

print("\nExpresado como fracciones:")
print("a) P(M) =", P_M_fraction)
print("b) P(H ∩ F) =", P_H_and_F_fraction)
print("c) P(F | M) =", P_F_given_M_fraction)
```

```python
#Problema 2

# Valores posibles de lambda y sus probabilidades a priori
lambdas = [1, 2, 3, 4]
prior_probabilities = [0.4, 0.3, 0.2, 0.1]

# Observación
x = 4  # Número de estudiantes observados

import numpy as np

# a) Estimación inicial del parámetro λ (lambda)
# λ̂ = E[λ] = Σ (λ_i * Π(λ_i))
initial_lambda_hat = sum(l * p for l, p in zip(lambdas, prior_probabilities))

print("a) La estimación inicial del parámetro λ̂ es:", initial_lambda_hat)

# b) Cálculo de la función de verosimilitud para cada modelo
# L(x | λ_i) = (e^{-λ_i} * λ_i^{x}) / x!
from math import exp, factorial

likelihoods = []
for l in lambdas:
    likelihood = (exp(-l) * l**x) / factorial(x)
    likelihoods.append(likelihood)

print("\nb) Función de verosimilitud para cada modelo:")
for i, (l, likelihood) in enumerate(zip(lambdas, likelihoods), 1):
    print(f"L(x=4 | λ={l}) = {likelihood:.6f}")

# c) Cálculo de la distribución a posteriori para cada modelo usando el teorema de Bayes
# P(λ_i | x) = [L(x | λ_i) * Π(λ_i)] / Σ_j [L(x | λ_j) * Π(λ_j)]

# Calculamos el numerador y denominador
numerators = [L * P for L, P in zip(likelihoods, prior_probabilities)]
denominator = sum(numerators)

posterior_probabilities = [num / denominator for num in numerators]

print("\nc) Distribución a posteriori para cada modelo:")
for l, posterior in zip(lambdas, posterior_probabilities):
    print(f"P(λ={l} | x=4) = {posterior:.4f}")

# d) Verificación de que la distribución posterior está normalizada
sum_posterior = sum(posterior_probabilities)
print(f"\nd) La suma de las probabilidades a posteriori es: {sum_posterior:.4f}")

# e) Modelo que más probablemente explica los datos
max_posterior = max(posterior_probabilities)
best_lambda = lambdas[posterior_probabilities.index(max_posterior)]

print(f"\ne) El modelo que más probablemente explica los datos es λ = {best_lambda}")

# f) Cálculo del mejor parámetro λ̂ usando la distribución a posteriori
# λ̂ = E[λ | x] = Σ (λ_i * P(λ_i | x))

lambda_hat_posterior = sum(l * p for l, p in zip(lambdas, posterior_probabilities))

print(f"\nf) El mejor parámetro λ̂ usando la distribución a posteriori es: {lambda_hat_posterior:.2f}")
```

```python
# Problema 3

# Valores posibles de lambda y sus probabilidades a priori
lambdas = [1, 2, 3, 4]
prior_probabilities = [0.4, 0.3, 0.2, 0.1]

# Observaciones
x1 = 4  # Número de estudiantes observados en el primer intervalo
x2 = 5  # Número de estudiantes observados en el segundo intervalo

import numpy as np

# a) Estimación inicial del parámetro λ (lambda)
# λ̂ = E[λ] = Σ (λ_i * Π(λ_i))
initial_lambda_hat = sum(l * p for l, p in zip(lambdas, prior_probabilities))

print("a) La estimación inicial del parámetro λ̂ es:", initial_lambda_hat)

# b) Cálculo de la función de verosimilitud conjunta para cada modelo
# L(x1, x2 | λ_i) = L(x1 | λ_i) * L(x2 | λ_i)
# L(x | λ_i) = (e^{-λ_i} * λ_i^{x}) / x!

from math import exp, factorial

def poisson_likelihood(x, l):
    return (exp(-l) * l**x) / factorial(x)

likelihoods_joint = []
for l in lambdas:
    L_x1 = poisson_likelihood(x1, l)
    L_x2 = poisson_likelihood(x2, l)
    likelihood_joint = L_x1 * L_x2
    likelihoods_joint.append(likelihood_joint)

print("\nb) Función de verosimilitud conjunta para cada modelo:")
for i, (l, likelihood) in enumerate(zip(lambdas, likelihoods_joint), 1):
    print(f"L(x1={x1}, x2={x2} | λ={l}) = {likelihood:.6e}")

# c) Cálculo de la distribución a posteriori para cada modelo usando el teorema de Bayes
# P(λ_i | x1, x2) = [L(x1, x2 | λ_i) * Π(λ_i)] / Σ_j [L(x1, x2 | λ_j) * Π(λ_j)]

# Calculamos el numerador y denominador
numerators = [L * P for L, P in zip(likelihoods_joint, prior_probabilities)]
denominator = sum(numerators)

posterior_probabilities = [num / denominator for num in numerators]

print("\nc) Distribución a posteriori para cada modelo:")
for l, posterior in zip(lambdas, posterior_probabilities):
    print(f"P(λ={l} | x1={x1}, x2={x2}) = {posterior:.4f}")

# d)
sum_posterior = sum(posterior_probabilities)
print(f"\nd) La suma de las probabilidades a posteriori es: {sum_posterior:.4f}")

# e) Modelo que más probablemente explica los datos
max_posterior = max(posterior_probabilities)
best_lambda = lambdas[posterior_probabilities.index(max_posterior)]

print(f"\ne) El modelo que más probablemente explica los datos es λ = {best_lambda}")

# f) Cálculo del mejor parámetro λ̂ usando la distribución a posteriori
# λ̂ = E[λ | x1, x2] = Σ (λ_i * P(λ_i | x1, x2))

lambda_hat_posterior = sum(l * p for l, p in zip(lambdas, posterior_probabilities))

print(f"\nf) El mejor parámetro λ̂ usando la distribución a posteriori es: {lambda_hat_posterior:.2f}")
```

# Generales de Probabilidad

```python
# Problema 1

from fractions import Fraction

# Generamos el espacio muestral de lanzar dos dados
total_outcomes = 6 * 6  # Total de combinaciones posibles con dos dados

# Evento A: La suma de los resultados es menor o igual a 3
event_A = []
for i in range(1, 7):
    for j in range(1, 7):
        if i + j <= 3:
            event_A.append((i, j))
P_A = Fraction(len(event_A), total_outcomes)

# Evento B: El resultado del primer lanzamiento es impar
event_B = []
for i in range(1, 7):
    if i % 2 != 0:  # Si el primer lanzamiento es impar
        for j in range(1, 7):
            event_B.append((i, j))
P_B = Fraction(len(event_B), total_outcomes)

# Evento A ∩ B: La suma es menor o igual a 3 y el primer lanzamiento es impar
event_A_and_B = [outcome for outcome in event_A if outcome in event_B]
P_A_and_B = Fraction(len(event_A_and_B), total_outcomes)

# Evento A ∪ B: La suma es menor o igual a 3 o el primer lanzamiento es impar
event_A_or_B = set(event_A + event_B)
P_A_or_B = Fraction(len(event_A_or_B), total_outcomes)

print("Problema 1:")
print(f"P(A) = {P_A} = {float(P_A)}")
print(f"P(B) = {P_B} = {float(P_B)}")
print(f"P(A ∩ B) = {P_A_and_B} = {float(P_A_and_B)}")
print(f"P(A ∪ B) = {P_A_or_B} = {float(P_A_or_B)}")
```

```python
# Problema 2

import math
from fractions import Fraction

# Función para calcular combinaciones
def combinaciones(n, k):
    return math.comb(n, k)

# Datos
total_equipos = 50
defectuosos = 2
no_defectuosos = total_equipos - defectuosos
equipos_muestra = 5

# Probabilidad de que ningún celular sea defectuoso
total_formas = combinaciones(total_equipos, equipos_muestra)
formas_no_defectuosos = combinaciones(no_defectuosos, equipos_muestra)

P_ningun_defectuoso = Fraction(formas_no_defectuosos, total_formas)

# Probabilidad de que al menos un celular sea defectuoso
P_al_menos_un_defectuoso = 1 - P_ningun_defectuoso

print("\nProblema 2:")
print(f"P(al menos un defectuoso) = {P_al_menos_un_defectuoso} = {float(P_al_menos_un_defectuoso)}")
```

```python
# Problema 3

# Datos
P_D = 0.6  # Suscrito al diario
P_C = 0.8  # Suscrito al cable
P_DyC = 0.5  # Suscrito a ambos servicios

# (a) Probabilidad de que esté suscrito a al menos uno de los dos servicios
P_D_o_C = P_D + P_C - P_DyC

print("\nProblema 3:")
print(f"(a) P(D ∪ C) = {P_D_o_C}")

# (b) Probabilidad de que esté suscrito a uno de los dos, pero no a ambos
# P(exactamente uno) = P(D ∪ C) - P(D ∩ C)
P_exactamente_uno = P_D_o_C - P_DyC

print(f"(b) P(suscrito a uno pero no a ambos) = {P_exactamente_uno}")
```

```python
# Problema 4

import matplotlib.pyplot as plt

# Lista para almacenar las probabilidades
probabilities = []

# Número total de días en un año
total_days = 365

# Calculamos P(n) para n desde 1 hasta 80
for n in range(1, 81):
    prob = 1.0
    for k in range(n):
        prob *= (total_days - k) / total_days
    probabilities.append(prob)

# Generamos la gráfica
n_values = range(1, 81)

plt.figure(figsize=(12, 6))
plt.plot(n_values, probabilities, marker='o')
plt.title('Probabilidad de que n personas tengan cumpleaños diferentes')
plt.xlabel('Número de personas (n)')
plt.ylabel('Probabilidad P(n)')
plt.grid(True)
plt.show()
```

# Hidden Markov Models

```python
# Solución al problema utilizando modelos ocultos de Markov

import numpy as np

# Definimos los estados ocultos y observables
estados = ['J', 'B']  # J: Moneda Justa, B: Moneda Sesgada
observaciones = ['C', 'S']  # C: Cara, S: Sello

# Matriz de transición T
T = np.array([[0.8, 0.2],
              [0.2, 0.8]])

# Matriz de emisión E
E = np.array([[0.5, 0.5],
              [0.9, 0.1]]) # Moneda Sesgada: P(C), P(S)

# Distribución inicial pi
pi = np.array([0.2, 0.8])  # [P(J), P(B)]

# Secuencia de observaciones
O_seq = ['S', 'C', 'C', 'C', 'S', 'C', 'S', 'C']
O = [observaciones.index(o) for o in O_seq]

# a) distribución a priori pi

print("a) Usando la distribución de probabilidad a priori pi = [0.2, 0.8] para la moneda justa y sesgada.")

# b) Encontrar la secuencia oculta más probable (Viterbi)

def viterbi(pi, T, E, O):
    N = len(pi)
    T_len = len(O)

    delta = np.zeros((T_len, N))
    psi = np.zeros((T_len, N), dtype=int)

    # Inicialización
    delta[0] = pi * E[:, O[0]]

    # Recursión
    for t in range(1, T_len):
        for j in range(N):
            temp = delta[t-1] * T[:, j]
            delta[t, j] = np.max(temp) * E[j, O[t]]
            psi[t, j] = np.argmax(temp)

    # Terminación
    P = np.max(delta[T_len - 1])
    path = np.zeros(T_len, dtype=int)
    path[T_len - 1] = np.argmax(delta[T_len - 1])

    # Retroceso
    for t in range(T_len - 2, -1, -1):
        path[t] = psi[t + 1, path[t + 1]]

    estado_optimo = [estados[state] for state in path]

    return estado_optimo, P

estado_optimo, P = viterbi(pi, T, E, O)

print("\nb) La secuencia oculta más probable es:")
print(estado_optimo)
print(f"Con una probabilidad de {P:.6f}")

# c)

def forward(pi, T, E, O):
    N = len(pi)
    T_len = len(O)

    alpha = np.zeros((T_len, N))

    # Inicialización
    alpha[0] = pi * E[:, O[0]]

    # Recursión
    for t in range(1, T_len):
        for j in range(N):
            alpha[t, j] = np.sum(alpha[t -1] * T[:, j]) * E[j, O[t]]
    P_o = np.sum(alpha[T_len - 1])

    return alpha, P_o

alpha, P_o = forward(pi, T, E, O)

print(f"\nc) La probabilidad de la secuencia observada es: {P_o:.6f}")

# d) Verificar que la suma de las probabilidades de todos los estados observables es 1

print("\nd) Verificación de que la suma de las probabilidades de todas las secuencias observables es aproximadamente 1.")

# Debido a la gran cantidad de posibles secuencias, en lugar de calcular todas, verificamos que las matrices están correctamente normalizadas
suma_T = np.allclose(T.sum(axis=1), 1)
suma_E = np.allclose(E.sum(axis=1), 1)
suma_pi = np.isclose(pi.sum(), 1)

if suma_T and suma_E and suma_pi:
    print("Las matrices de transición y emisión, y la distribución inicial están correctamente normalizadas.")
    print("Por lo tanto, la suma de las probabilidades de todas las secuencias observables es 1.")
else:
    print("Hay un error en la normalización de las matrices o la distribución inicial.")

# e) ¿Depende el resultado de la probabilidad a priori?

print("\ne) Sí, el resultado depende de la probabilidad a priori.")
print("Si cambiamos la distribución inicial pi, las probabilidades calculadas en los apartados b) y c) pueden cambiar.")
```

```python
#Ejercicio 6
import numpy as np

# Número de puntos aleatorios
N = 1000000

# Generación de radios aleatorios
u = np.random.uniform(0, 1, N)
r = u ** (1/3)  # Raíz cúbica para distribución uniforme en volumen

# Generación de ángulos aleatorios
theta = np.random.uniform(0, 2*np.pi, N)
phi = np.arccos(1 - 2 * np.random.uniform(0, 1, N))  # Distribución uniforme en cos(phi)

# Conversión a coordenadas cartesianas
x_array = r * np.sin(phi) * np.cos(theta)
y_array = r * np.sin(phi) * np.sin(theta)
z_array = r * np.cos(phi)

# Cálculo de los momentos de inercia
I_xx = np.mean(y_array**2 + z_array**2)
I_yy = np.mean(x_array**2 + z_array**2)
I_zz = np.mean(x_array**2 + y_array**2)

# Cálculo del producto de inercia
I_xy = -np.mean(x_array * y_array)

# Resultados
print(f"Estimación de I_xx: {I_xx:.6f}")
print(f"Estimación de I_yy: {I_yy:.6f}")
print(f"Estimación de I_zz: {I_zz:.6f}")
print(f"Estimación de I_xy: {I_xy:.6f}")

# Valor exacto
I_exacto = 2 / 5  # 0.4

# Errores
error_I_xx = abs(I_xx - I_exacto)
error_I_yy = abs(I_yy - I_exacto)
error_I_zz = abs(I_zz - I_exacto)

print(f"\nValor exacto de I: {I_exacto}")
print(f"Error en I_xx: {error_I_xx:.6f}")
print(f"Error en I_yy: {error_I_yy:.6f}")
print(f"Error en I_zz: {error_I_zz:.6f}")

# Conclusión sobre la simetría
print("\nComo la esfera es simétrica, los productos de inercia (e.g., I_xy) deben ser cero.")
print(f"Estimación de I_xy: {I_xy:.6f}")

# Mostrar el texto en la ejecución
print("Los resultados confirman que la esfera posee simetría de rotación perfecta. Sus propiedades físicas no cambian bajo rotaciones alrededor de cualquier eje que pase por su centro.")

```

```python
#Ejercicio 9

import numpy as np
import matplotlib.pyplot as plt

# Define the Beta(2,4) PDF
def beta_pdf(x):
    return 20 * x * (1 - x)**3

# Find the maximum value of the function
from scipy.optimize import minimize_scalar

result = minimize_scalar(lambda x: -beta_pdf(x), bounds=(0, 1), method='bounded')
f_max = beta_pdf(result.x)

print(f"Maximum value of f(x) is approximately {f_max:.4f} at x = {result.x:.4f}")

# Acceptance-Rejection Method
def acceptance_rejection(N):
    # Generate N random x values between 0 and 1
    x = np.random.uniform(0, 1, N)
    # Generate N random y values between 0 and f_max
    y = np.random.uniform(0, f_max, N)
    # Count the number of acceptances
    accept = y <= beta_pdf(x)
    num_accept = np.sum(accept)
    # Estimated area under the curve
    area_estimate = num_accept / N * (1 * f_max)  # Width * Height
    # Standard deviation (uncertainty)
    p = num_accept / N
    sigma = np.sqrt(p * (1 - p) / N) * (1 * f_max)
    # Relative uncertainty
    relative_uncertainty = sigma / area_estimate
    return area_estimate, sigma, relative_uncertainty

# Determine N required for 1% uncertainty
# We can start with an initial N and increase until the relative uncertainty is <= 0.01
N = 1000  # Initial number of samples
relative_uncertainty = 1.0  # Start with 100% uncertainty

while relative_uncertainty > 0.01:
    area_estimate, sigma, relative_uncertainty = acceptance_rejection(N)
    print(f"N = {N}, Estimated Area = {area_estimate:.5f}, Relative Uncertainty = {relative_uncertainty:.5f}")
    if relative_uncertainty > 0.01:
        N *= 2

print("\nFinal Result:")
print(f"N = {N}")
print(f"Estimated Area = {area_estimate:.5f}")
print(f"Standard Deviation = {sigma:.5f}")
print(f"Relative Uncertainty = {relative_uncertainty * 100:.2f}%")

# Generate points for plotting the function
x_plot = np.linspace(0, 1, 500)
y_plot = beta_pdf(x_plot)

# Generate the accepted and rejected points for plotting
x = np.random.uniform(0, 1, N)
y = np.random.uniform(0, f_max, N)
accept = y <= beta_pdf(x)

plt.figure(figsize=(10, 6))
plt.plot(x_plot, y_plot, label='Beta(2,4) PDF')
plt.fill_between(x_plot, y_plot, alpha=0.2)
plt.scatter(x[accept], y[accept], color='green', s=1, label='Accepted Points')
plt.scatter(x[~accept], y[~accept], color='red', s=1, label='Rejected Points')
plt.title('Acceptance-Rejection Sampling for Beta(2,4) Distribution')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```
