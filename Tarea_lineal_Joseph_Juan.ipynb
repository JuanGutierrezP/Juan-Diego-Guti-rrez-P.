```python
#Punto 2
import numpy as np

# sistema
A = np.array([[3., -1., -1.],
              [-1., 3., 1.],
              [2., 1., 4.]])
b = np.array([1., 3., 7.])
def gauss_seidel(A, b, x0=None, tol=1e-10, max_iter=1000):
    n = len(b)
    if x0 is None:
        x = np.zeros(n)
    else:
        x = x0.copy()
    for k in range(max_iter):
        x_old = x.copy()
        for i in range(n):
            sum_ = 0
            for j in range(n):
                if j != i:
                    sum_ += A[i, j] * x[j]
            x[i] = (b[i] - sum_) / A[i, i]
        if np.linalg.norm(x - x_old, ord=np.inf) < tol:
            print(f"Convergencia hecha en {k+1} iteraciones.")
            return x
    print("No se alcanzó la convergencia después del número máximo de iteraciones posibles")
    return x

x0 = np.zeros(len(b))
x_sol = gauss_seidel(A, b, x0)
print("Solución aproximada:")
print(x_sol)

# Verificar la solución
residual = np.dot(A, x_sol) - b
print("Residual:")
print(residual)
```

```python
#Punto 3
import numpy as np
class LinearSolver:
    def __init__(self, A, b):
        self.A = np.array(A, dtype=float)
        self.b = np.array(b, dtype=float)
        self.n = len(b)
    def jacobi(self, x0=None, tol=1e-10, max_iterations=1000):
        A = self.A
        b = self.b
        n = self.n
        x = np.zeros(n) if x0 is None else x0
        D = np.diag(np.diag(A))
        R = A - D
        iterations = 0
        for i in range(max_iterations):
            x_new = (b - np.dot(R, x)) / np.diag(A)
            iterations += 1

            if np.linalg.norm(x_new - x, ord=np.inf) < tol:
                print(f"Método de Jacobi convergió en {iterations} iteraciones.")
                return x_new
            x = x_new
        print(f"Método de Jacobi no convergió en {max_iterations} iteraciones.")
        return x
    def gauss_seidel(self, x0=None, tol=1e-10, max_iterations=1000):
        A = self.A
        b = self.b
        n = self.n
        x = np.zeros(n) if x0 is None else x0
        iterations = 0
        for k in range(max_iterations):
            x_new = np.copy(x)
            for i in range(n):
                # s1 y s2 para sumar los productos anteriores
                s1 = np.dot(A[i, :i], x_new[:i])
                s2 = np.dot(A[i, i+1:], x[i+1:])
                x_new[i] = (b[i] - s1 - s2) / A[i, i]
            iterations += 1
            # Verificar la convergencia
            if np.linalg.norm(x_new - x, ord=np.inf) < tol:
                print(f"Método de Gauss-Seidel convergió en {iterations} iteraciones.")
                return x_new
            x = x_new
        print(f"Método de Gauss-Seidel no convergió en {max_iterations} iteraciones.")
        return x

A = [[10, 2, -1],[-3, -6, 2],[1, 1, 5]]
b = [27, -61.5, -21.5]
solver = LinearSolver(A, b)
# método de Jacobi
x_jacobi = solver.jacobi()
print("Solución por Jacobi:", x_jacobi)
# Gauss-Seidel
x_gauss_seidel = solver.gauss_seidel()
print("Solución por Gauss-Seidel:", x_gauss_seidel)
```

```python
#Punto 4

def matriz_multi(A, B):
    # Obtener las dimensiones de las matrices A y B
    filas_A = len(A)
    columnas_A = len(A[0])
    filas_B = len(B)
    columnas_B = len(B[0])

    if columnas_A != filas_B:
        raise ValueError("No se pueden multiplicar las matrices: son dimensiones incompatibles.")

    #matriz de ceros
    res = [[0 for _ in range(columnas_B)] for _ in range(filas_A)]

    # multiplicación de matrices
    for i in range(filas_A):
        for j in range(columnas_B):
            for k in range(columnas_A):
                res[i][j] += A[i][k] * B[k][j]

    return res

# Definir las matrices A y B
A = [
    [1, 0, 0],
    [5, 1, 0],
    [-2, 3, 1]]

B = [
    [4, -2, 1],
    [0, 3, 7],
    [0, 0, 2]]

# Calcular el producto AB
AB = matriz_multi(A, B)
# Imprimir el resultado
print("El producto de las matrices A y B es :")
for fila in AB:
    print(fila)
```

```python
#Punto 8
import numpy as np
def lu_decomposition(A):
    n = A.shape[0]
    L = np.zeros((n, n))
    U = np.zeros((n, n))

    for i in range(n):
        # Construcción de U
        for k in range(i, n):
            sum = 0
            for j in range(i):
                sum += L[i][j] * U[j][k]
            U[i][k] = A[i][k] - sum

        # Construcción de L
        for k in range(i, n):
            if i == k:
                L[i][i] = 1  # Diagonal de L es 1
            else:
                sum = 0
                for j in range(i):
                    sum += L[k][j] * U[j][i]
                L[k][i] = (A[k][i] - sum) / U[i][i]
    return L, U
# Definir la matriz A
A = np.array([
    [4, -2, 1],
    [20, -7, 12],
    [-8, 13, 17]
], dtype=float)

# Realizar la descomposición LU
L, U = lu_decomposition(A)

# Mostrar los resultados
print("Matriz L:")
print(L)
print("\nMatriz U:")
print(U)

# Verificar que A = L * U
A_reconstructed = np.dot(L, U)
print("\n(A reconstruida):")
print(A_reconstructed)
```

```python
#Punto 10
import numpy as np

#a)
def jacobi_eigenvalue(A, tol=1e-10, max_iterations=100):
    n = A.shape[0]
    A = A.copy()
    V = np.eye(n)
    iterations = 0

    for _ in range(max_iterations):
        # Encontrar el elemento fuera de la diagonal más grande en valor absoluto
        off_diag = np.abs(A - np.diag(np.diagonal(A)))
        i, j = np.unravel_index(np.argmax(off_diag), A.shape)
        if A[i, j] == 0 or np.abs(A[i, j]) < tol:
            break

        # Calcular el ángulo theta
        if A[i, i] == A[j, j]:
            theta = np.pi / 4
        else:
            theta = 0.5 * np.arctan2(2 * A[i, j], A[i, i] - A[j, j])

        # Construir la matriz de rotación
        R = np.eye(n)
        R[i, i] = np.cos(theta)
        R[j, j] = np.cos(theta)
        R[i, j] = -np.sin(theta)
        R[j, i] = np.sin(theta)

        # Actualizar A y V
        A = R.T @ A @ R
        V = V @ R
        iterations += 1

    # Los valores propios son los elementos diagonales de A
    eigenvalues = np.diagonal(A)
    # Los vectores propios son las columnas de V
    Propiovectors = V

    return eigenvalues, Propiovectors, iterations

# Matriz A proporcionada
A = np.array([
    [4, 1, 1],
    [1, 3, 2],
    [1, 2, 5]
], dtype=float)

# Ejecutar el método de Jacobi
eigenvalues, Propiovectors, iterations = jacobi_eigenvalue(A)

# Ordenar los valores y vectores propios
idx = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[idx]
Propiovectors = Propiovectors[:, idx]

# Mostrar los resultados
print(f"Número de iteraciones: {iterations}\n")
print("Valores propios obtenidos por el método de Jacobi:")
print(eigenvalues)
print("\nVectores propios correspondientes:")
print(Propiovectors)

eigvals_numpy, eigvecs_numpy = np.linalg.eig(A)
# Ordenar los valores propios de NumPy
idx_numpy = np.argsort(eigvals_numpy)[::-1]
eigvals_numpy = eigvals_numpy[idx_numpy]
eigvecs_numpy = eigvecs_numpy[:, idx_numpy]

print("\nValores propios de numpy:")
print(eigvals_numpy)
print("\nVectores propios correspondientes de numpy:")
print(eigvecs_numpy)
```

```python
#Punto 13

import numpy as np

def jacobian_order4(F, x, h=1e-5):
    n = len(x)
    Fx = F(x)
    m = len(Fx)
    J = np.zeros((m, n))
    h_vec = np.zeros_like(x)

    for i in range(n):
        h_vec[i] = h
        f_plus_2h = F(x + 2*h_vec)
        f_plus_h = F(x + h_vec)
        f_minus_h = F(x - h_vec)
        f_minus_2h = F(x - 2*h_vec)
        J[:, i] = (-f_plus_2h + 8*f_plus_h - 8*f_minus_h + f_minus_2h) / (12*h)
        h_vec[i] = 0

    return J

def jacobian_order2(F, x, h=1e-5):
    n = len(x)
    Fx = F(x)
    m = len(Fx)
    J = np.zeros((m, n))
    h_vec = np.zeros_like(x)

    for i in range(n):
        h_vec[i] = h
        f_plus_h = F(x + h_vec)
        f_minus_h = F(x - h_vec)
        J[:, i] = (f_plus_h - f_minus_h) / (2*h)
        h_vec[i] = 0

    return J

def F_system(x):
    x1, x2, x3 = x
    f1 = 6*x1 - 2*np.cos(x2 * x3) - 1
    f2 = 9*x2 + np.sqrt(x1**2 + np.sin(x3) + 1.06) + 0.9
    f3 = 60*x3 + 3*np.exp(-x1 * x2) + 10*np.pi - 3
    return np.array([f1, f2, f3])


x_point = np.array([0.5, 0.5, 0.5])

h = 0.01

# Calcular Jacobiano con(h^4)
J_order4 = jacobian_order4(F_system, x_point, h)
print("Jacobiano con operador de orden O(h^4):")
print(J_order4)

# Calcular Jacobiano con (h^2)
J_order2 = jacobian_order2(F_system, x_point, h)
print("\nJacobiano con operador de orden O(h^2):")
print(J_order2)

#h = 0.0001 para igualar precisiones
h_new = 0.0001
J_order2_new = jacobian_order2(F_system, x_point, h_new)
print("\nJacobiano con operador de orden O(h^2) y h=0.0001:")
print(J_order2_new)
```

```python
#Punto 14
import sympy as sp

# Definir símbolos
x, y, z = sp.symbols('x y z')

# Generadores del grupo SO(3)
J_x = sp.Matrix([
    [0,  0,  0],
    [0,  0, -1],
    [0,  1,  0]
])

J_y = sp.Matrix([
    [0,  0,  1],
    [0,  0,  0],
    [-1, 0,  0]
])

J_z = sp.Matrix([
    [0, -1,  0],
    [1,  0,  0],
    [0,  0,  0]
])

# símbolo de Levi-Civita
def levi_civita(i, j, k):
    if (i, j, k) in [(1, 2, 3), (2, 3, 1), (3, 1, 2)]:
        return 1
    elif (i, j, k) in [(3, 2, 1), (2, 1, 3), (1, 3, 2)]:
        return -1
    else:
        return 0

# generadores
generators = [J_x, J_y, J_z]
generador_cu = ['J_x', 'J_y', 'J_z']

def print_commutator(name1, name2, result):
    print(f"\nConmutador [{name1}, {name2}]:")
    sp.pprint(result)

#  conmutadores
for i in range(3):
    for j in range(3):
        if i < j:
            Ji = generators[i]
            Jj = generators[j]
            comm = Ji * Jj - Jj * Ji
            # Encontrar k tal que (i, j, k) es una permutación de (0, 1, 2)
            k = 3 - i - j
            epsi = levi_civita(i+1, j+1, k+1)
            expected = epsi * generators[k]
            print_commutator(generador_cu[i], generador_cu[j], comm)
            print(f"¿[{generador_cu[i]}, {generador_cu[j]}] = ε{i+1}{j+1}{k+1} * {generador_cu[k]}?")
            sp.pprint(comm.equals(expected))
```
