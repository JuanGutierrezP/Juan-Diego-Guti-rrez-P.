```python
#Ejercicio 2.0.2
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# Descargar los datos
url = "https://raw.githubusercontent.com/asegura4488/Database/main/MetodosComputacionalesReforma/EstrellaEspectro.txt"
data = np.loadtxt(url)

# Encontrar los máximos locales
peaks, _ = find_peaks(data)

# Graficar los datos y los máximos
plt.figure(figsize=(20, 10))
plt.plot(data, label="Datos 1")
plt.plot(peaks, data[peaks], "r", label=" Máximos locales")
plt.title(" máximos locales")
plt.xlabel("Índice")
plt.ylabel("Valor")
plt.legend()
plt.show()
```

```python
#Ejercicio 2
import numpy as np
import matplotlib.pyplot as plt
def f(x):
  return 1/(np.sqrt(1+np.e**(-x**2)))
x= np.linspace(-10,10,500)
y=f(x)

plt.scatter(x,y,color="b")
plt.title("función ")
plt.show()
#---
def derivada_central(f,x,h=0.05):
  d=0
  if h!=0:
    d= (f(x+h)-f(x-h))/(2*h)
  return d
y=derivada_central(f,x,h=0.05)
x=np.linspace(-10,10,500)
plt.scatter(x,y,color="r")
plt.title("Derivada central")
plt.show()

```

```python
#Ejercicio 6 4.34
import numpy as np

x = np.linspace(-4, 4, 25)
y = np.linspace(-4, 4, 25)
X, Y = np.meshgrid(x, y)
#b
def potencia_flujo(x, y):
  V = 2
  R = 2
  return V * x *  (1 - R**2 / (x**2 + y**2))
#c
h = 0.001

V_x = np.zeros_like(X)
V_y = np.zeros_like(X)
for i in range(1, len(x)-1):
  for j in range(1, len(y)-1):

     V_x[i, j] = (potencia_flujo(X[i,j]+h, Y[i,j]) - potencia_flujo(X[i,j]-h, Y[i,j])) /(2*h)
     V_y[i, j] = -(potencia_flujo(X[i,j], Y[i,j]+h) - potencia_flujo(X[i,j], Y[i,j]-h)) /(2*h)
     V_x[X**2+Y**2<4]=0
     V_y[X**2+Y**2<4]=0

#d
import matplotlib.pyplot as plt
fig,ax=plt.subplots(figsize=(9,9))

ax.quiver(X, Y, V_x, V_y, color= "b")
ax.set_xlabel('x[Cm]')
ax.set_ylabel('y[Cm]')
ax.set_title('Velocidad cerca al cilindro ')
c=plt.Circle((0,0),2,color="red", fill= False)
ax.add_patch(c)
plt.show()

```
