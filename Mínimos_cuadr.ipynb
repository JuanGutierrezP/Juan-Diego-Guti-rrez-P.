```python
#Cap 9 M√≠nimos cuadrados
#Punto 1
#A
import numpy as np
import matplotlib.pyplot as plt

# Definir la matriz A y el vector b
A = np.array([[2, -1],
              [2, 2],
              [1, 1]])

b = np.array([0, 1, 4])

# Calcular el punto de m√≠nimos cuadrados
x_hat = np.linalg.inv(A.T @ A) @ A.T @ b

x_sol, y_sol = x_hat

# Mostrar el punto soluci√≥n
print(f"Punto soluci√≥n: (x, y) = ({x_sol:.2f}, {y_sol:.2f})")

# Graficar las l√≠neas
x = np.linspace(-1, 5, 100)

# Graficar cada l√≠nea
plt.plot(x, 2*x, label='2x - y = 0')
plt.plot(x, 0.5 - x, label='2x + 2y = 1')
plt.plot(x, 4 - x, label='x + y = 4')

# Graficar el punto soluci√≥n
plt.plot(x_sol, y_sol, 'ro', label='Punto soluci√≥n')

# Configurar la gr√°fica
plt.title('Gr√°fico de las tres l√≠neas y el punto soluci√≥n')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.grid()
plt.legend()
plt.show()
#B
x_range = np.arange(-5, 5, 0.01)
y_range = np.arange(-5, 5, 0.01)

# Inicializar la distancia m√≠nima y las coordenadas del punto m√°s cercano
min_dist = float('inf')
x_min, y_min = 0, 0

# Iterar sobre todos los puntos en el rango de b√∫squeda
for x in x_range:
    for y in y_range:
        # Calcular la distancia a cada l√≠nea
        dist1 = abs(2*x - y) / np.sqrt(2**2 + (-1)**2)
        dist2 = abs(2*x + 2*y - 1) / np.sqrt(2**2 + 2**2)
        dist3 = abs(x + y - 4) / np.sqrt(1**2 + 1**2)

        # Calcular la distancia total
        dist = dist1 + dist2 + dist3

        # Actualizar la distancia m√≠nima y las coordenadas si es necesario
        if dist < min_dist:
            min_dist = dist
            x_min = x
            y_min = y

# Mostrar la distancia m√≠nima y las coordenadas del punto m√°s cercano
print(f"Distancia m√≠nima: {min_dist:.2f}")
print(f"Punto m√°s cercano: (x, y) = ({x_min:.2f}, {y_min:.2f})")
```

```python
#Punto 4
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Definir las funciones de ajuste
def func_lineal(x, a0, a1):
    return a0 + a1 * x

def func_cuadratica(x, a0, a1, a2):
    return a0 + a1 * x + a2 * x**2

# Cargar los datos para el ajuste lineal
url_lineal = 'https://raw.githubusercontent.com/asegura4488/Database/main/MetodosComputacionalesReforma/MinimosLineal.txt'
data_lineal = np.loadtxt(url_lineal)
x_lineal = data_lineal[:, 0]
y_lineal = data_lineal[:, 1]

# Cargar los datos para el ajuste cuadr√°tico
url_cuadratico = 'https://raw.githubusercontent.com/asegura4488/Database/main/MetodosComputacionalesReforma/MinimosCuadratico.txt'
data_cuadratico = np.loadtxt(url_cuadratico)
x_cuadratico = data_cuadratico[:, 0]
y_cuadratico = data_cuadratico[:, 1]

# Ajuste lineal con curve_fit
params_lineal, cov_lineal = curve_fit(func_lineal, x_lineal, y_lineal)
a0_lineal, a1_lineal = params_lineal

# Ajuste cuadr√°tico con curve_fit
params_cuadratico, cov_cuadratico = curve_fit(func_cuadratica, x_cuadratico, y_cuadratico)
a0_cuadratico, a1_cuadratico, a2_cuadratico = params_cuadratico

# Calcular los valores ajustados
y_fit_lineal = func_lineal(x_lineal, a0_lineal, a1_lineal)
y_fit_cuadratico = func_cuadratica(x_cuadratico, a0_cuadratico, a1_cuadratico, a2_cuadratico)

# Graficar el ajuste lineal
plt.figure(figsize=(8, 6))
plt.scatter(x_lineal, y_lineal, label='Datos')
plt.plot(x_lineal, y_fit_lineal, color='red', label=f'Ajuste curve_fit: $f(x) = {a0_lineal:.2f} + {a1_lineal:.2f} x$')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ajuste Lineal usando curve_fit')
plt.legend()
plt.grid(True)
plt.show()

# Graficar el ajuste cuadr√°tico
plt.figure(figsize=(8, 6))
plt.scatter(x_cuadratico, y_cuadratico, label='Datos')
plt.plot(x_cuadratico, y_fit_cuadratico, color='red', label=f'Ajuste curve_fit: $f(x) = {a0_cuadratico:.2f} + {a1_cuadratico:.2f} x + {a2_cuadratico:.2f} x^2$')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ajuste Cuadr√°tico usando curve_fit')
plt.legend()
plt.grid(True)
plt.show()

# Comparaci√≥n de resultados

# Ajuste lineal con np.polyfit
coef_polyfit_lineal = np.polyfit(x_lineal, y_lineal, 1)
a1_polyfit_lineal, a0_polyfit_lineal = coef_polyfit_lineal

print("Ajuste Lineal")
print("-------------")
print(f"Con np.polyfit: a0 = {a0_polyfit_lineal:.4f}, a1 = {a1_polyfit_lineal:.4f}")
print(f"Con curve_fit:  a0 = {a0_lineal:.4f}, a1 = {a1_lineal:.4f}")

# Ajuste cuadr√°tico con np.polyfit
coef_polyfit_cuadratico = np.polyfit(x_cuadratico, y_cuadratico, 2)
a2_polyfit_cuadratico, a1_polyfit_cuadratico, a0_polyfit_cuadratico = coef_polyfit_cuadratico

print("\nAjuste Cuadr√°tico")
print("-----------------")
print(f"Con np.polyfit: a0 = {a0_polyfit_cuadratico:.4f}, a1 = {a1_polyfit_cuadratico:.4f}, a2 = {a2_polyfit_cuadratico:.4f}")
print(f"Con curve_fit:  a0 = {a0_cuadratico:.4f}, a1 = {a1_cuadratico:.4f}, a2 = {a2_cuadratico:.4f}")
```

```python
#Punto 7
import numpy as np

# Definir los vectores base del subespacio W
u1 = np.array([3, 1, 0, 1])
u2 = np.array([1, 2, 1, 1])
u3 = np.array([-1, 0, 2, -1])

# Definir el vector objetivo b
b = np.array([-3, -3, 8, 9])

# Parte a): Proyecci√≥n usando m√≠nimos cuadrados matriciales

# Construir la matriz A con las columnas u1, u2 y u3
A = np.column_stack((u1, u2, u3))

# Resolver Ax = b usando m√≠nimos cuadrados
x, residuals, rank, s = np.linalg.lstsq(A, b, rcond=None)

# Calcular la proyecci√≥n p_W(b) = A x
p_b = A @ x

print("Coeficientes x:")
print(x)
print("\nProyecci√≥n p_W(b):")
print(p_b)

# Parte b): Proyecci√≥n usando el proceso de Gram-Schmidt

def gram_schmidt(vectors):
    orthonormal_basis = []
    for v in vectors:
        # Copiar el vector actual
        u = v.copy().astype(float)
        for b in orthonormal_basis:
            # Proyectar v sobre b y restar
            proj = np.dot(u, b) * b
            u -= proj
        # Normalizar u
        norm = np.linalg.norm(u)
        if norm > 1e-10:
            u /= norm
            orthonormal_basis.append(u)
    return orthonormal_basis

# Lista de vectores originales
vectors = [u1, u2, u3]

# Obtener la base ortonormal
orthonormal_basis = gram_schmidt(vectors)

# Calcular los coeficientes c_i = <b, v_i>
coefficients = []
for v in orthonormal_basis:
    c = np.dot(b, v)
    coefficients.append(c)

# Calcular p_W(b) = suma de c_i * v_i
p_b_gram_schmidt = sum(c * v for c, v in zip(coefficients, orthonormal_basis))

print("\nBase ortonormal (vectores v_i):")
for i, v in enumerate(orthonormal_basis, 1):
    print(f"v_{i} =", v)

print("\nCoeficientes c_i:")
for i, c in enumerate(coefficients, 1):
    print(f"c_{i} = {c}")

print("\nProyecci√≥n p_W(b) usando Gram-Schmidt:")
print(p_b_gram_schmidt)

# Verificaci√≥n final
print("\n¬øLas proyecciones son iguales?", np.allclose(p_b, p_b_gram_schmidt))
```

```python
#Bono

import numpy as np
import random
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def LennardJones(r, epsilon=1, sigma=1):
    """
    Potencial de Lennard-Jones para dos part√≠culas a una distancia r.
    """
    return 4 * epsilon * ((sigma / r)**12 - (sigma / r)**6)

def TotalEnergy(positions, epsilon=1, sigma=1):
    """
    Calcula la energ√≠a total de un sistema de part√≠culas en posiciones dadas.
    """
    N = len(positions)
    energy = 0
    for i in range(N):
        for j in range(i + 1, N):
            r = np.linalg.norm(positions[i] - positions[j])  # Distancia entre part√≠culas
            if r != 0:
                energy += LennardJones(r, epsilon, sigma)
    return energy

# Par√°metros del algoritmo gen√©tico
N_particles = 50  # N√∫mero de part√≠culas
population_size = 50  # Tama√±o de la poblaci√≥n
num_generations = 100  # N√∫mero de generaciones
mutation_rate = 0.1  # Tasa de mutaci√≥n
selection_size = 20  # N√∫mero de individuos seleccionados para reproducci√≥n

def create_individual():-
    """
    Crea un individuo con posiciones aleatorias para N part√≠culas en un cubo de lado L.
    """
    L = 10  # Tama√±o del cubo
    positions = np.random.uniform(0, L, (N_particles, 3))
    return positions

def create_population(size):
    """
    Crea una poblaci√≥n de individuos.
    """
    return [create_individual() for _ in range(size)]

def fitness(individual):
    """
    Calcula la aptitud de un individuo. En este caso, es la energ√≠a total (negativa para minimizar).
    """
    energy = TotalEnergy(individual)
    return -energy  # Negativo porque el algoritmo gen√©tico maximiza la aptitud

def selection(population, fitnesses, k):
    """
    Selecciona k individuos de la poblaci√≥n usando selecci√≥n por torneo.
    """
    selected = []
    for _ in range(k):
        i, j = random.sample(range(len(population)), 2)
        if fitnesses[i] > fitnesses[j]:
            selected.append(population[i])
        else:
            selected.append(population[j])
    return selected

def crossover(parent1, parent2):
    """
    Realiza cruzamiento uniforme entre dos padres.
    """
    mask = np.random.rand(N_particles, 3) < 0.5
    child = np.where(mask, parent1, parent2)
    return child

def mutate(individual, mutation_rate):
    """
    Aplica mutaci√≥n al individuo.
    """
    mutation = np.random.randn(N_particles, 3) * mutation_rate
    individual += mutation
    return individual

# Crear la poblaci√≥n inicial
population = create_population(population_size)

# Registrar el mejor individuo
best_individual = None
best_fitness = -np.inf
best_energies = []

for generation in range(num_generations):
    # Calcular las aptitudes
    fitnesses = [fitness(ind) for ind in population]

    # Actualizar el mejor individuo
    max_fitness = max(fitnesses)
    if max_fitness > best_fitness:
        best_fitness = max_fitness
        best_individual = population[fitnesses.index(max_fitness)]

    best_energies.append(-best_fitness)

    # Mostrar progreso cada 10 generaciones
    if generation % 10 == 0:
        print(f"Generaci√≥n {generation}, Mejor energ√≠a: {-best_fitness:.4f}")

    # Selecci√≥n
    selected = selection(population, fitnesses, selection_size)

    # Crear la nueva poblaci√≥n
    new_population = []
    while len(new_population) < population_size:
        parent1, parent2 = random.sample(selected, 2)
        child = crossover(parent1, parent2)
        child = mutate(child, mutation_rate)
        new_population.append(child)

    population = new_population

# Gr√°fica de convergencia
plt.plot(best_energies)
plt.xlabel('Generaci√≥n')
plt.ylabel('Mejor energ√≠a')
plt.title('Convergencia del algoritmo gen√©tico')
plt.show()

# Visualizaci√≥n de la configuraci√≥n final
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
positions = best_individual
ax.scatter(positions[:, 0], positions[:, 1], positions[:, 2])
ax.set_title('Configuraci√≥n de part√≠culas de menor energ√≠a')
plt.show()
```
